```python
import pandas as pd
from pathlib import Path
from shutil import rmtree
from pprint import pprint
```

```python
root = Path.cwd().parent
data = root / 'data' / 'raw'
ds2018 = data / '2018'
tv = ds2018 / 'train_val2018'
test = ds2018 / 'test2018'
ds2019 = data / '2019'
t = ds2019 / 'train_val2019'
ds2021 = data / '2021'
train = ds2021 / 'train'
v = ds2021 / 'val'
tm = ds2021 / 'train_mini'
```

```python
df = pd.read_json(data / '2018' / 'categories.json')
df = df[df['kingdom'] == 'Fungi']
df
```

```python
def filter_dirs(path):
    for p in path.iterdir():
        if p.name.startswith('Fungi'):
            print(p)
            continue
        else:
            rmtree(p)


def rename_dirs(path, target_dir):
    for p in path.iterdir():
        if not p.name.startswith("Fungi"):
            newname = '_'.join(p.name.split('_')[1:])
            if not (target_dir / newname).is_dir():
                (target_dir / newname).mkdir()
            for f in p.iterdir():
                f.rename(target_dir / newname / f.name)
            rmtree(p)
```

```python
map(filter_dirs, [tv, test, t, train,])

map(rename_dirs, [(v, train), (tm, train)])
```
