```python
def rename_images(fpath, label):
    for fname in os.listdir(fpath):
        # new_name = re.sub('\.\d+', '', fname)
        # new_name = str(num)+new_name
        new_name = label + fname
        os.rename(os.path.join(fpath, fname), os.path.join(fpath, new_name))
```

```python
def return_labeled_df(filepath, cloud_path, label):
    dir_list = []
    for fname in os.listdir(filepath):
        new = cloud_path + fname
        dir_list.append(new)
    df = pd.DataFrame(dir_list)
    df['label'] = label
    return df
```

```python
ascos = return_labeled_df(filepath='/content/gcp/ascos', cloud_path='g://auto_ml_muchrooms/ascos/', label='None_of_the_above')
basidios = return_labeled_df(filepath='/content/gcp/basidios', cloud_path='g://auto_ml_muchrooms/basidios/', label='None_of_the_above')
morels = return_labeled_df(filepath='/content/gcp/morels/inat', cloud_path='g://auto_ml_muchrooms/morels/inat/', label='Morels')
gyro = return_labeled_df(filepath='/content/gcp/gyro/inat', cloud_path='g://auto_ml_muchrooms/gyro/inat/', label='Gyromitra')
```
